# Now Training the model

```python
finalData = data.drop(columns="Sales")
targetedData = data["Sales"]
```

```python
XTrain, XTest, YTrain, YTest = train_test_split(finalData, targetedData, test_size=.2, random_state=42)
```

```python
regressor = LinearRegression()
```

```python
regressor.fit(XTrain,YTrain)
```

```python
predection = regressor.predict(XTest)
predection.shape
```

```python
mae = mean_absolute_error(YTest, predection)
print("mean absolute error: ",mae)
```
